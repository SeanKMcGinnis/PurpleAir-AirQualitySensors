```python
import re
import requests
from arcgis.gis import GIS
gis = GIS("home")
## Script Specific Variables
# URL of PurpleAir API
api_url = "https://api.purpleair.com/v1/sensors"
# PurpleAir API Key
api_key = ""
# Data Fields Returned in Response
fields = "sensor_index,name,location_type,latitude,longitude,altitude,last_seen,channel_state,channel_flags,confidence,pm1.0,pm2.5,pm2.5_10minute,pm2.5_30minute,pm2.5_60minute,pm2.5_6hour,pm2.5_24hour,pm2.5_1week,pm10.0,0.3_um_count,0.5_um_count,1.0_um_count,2.5_um_count,5.0_um_count,10.0_um_count,humidity,temperature,pressure,voc,ozone1,analog_input"
# ArcGIS Online Item Id's
# Dynamic Hosted feature layer item ID
fc_id = ""
#Archive Hosted feature layer item ID
archive_id = ""
```

```python
def none_handler(val):
    if(val == None):
        val = 0
    return val
```

```python
def string_handler(val):
    val = re.sub(r"[^a-zA-Z0-9]","",val)
    return val
```

```python
## Get the GIS Objects
sensor_service = gis.content.get(fc_id)
sensor_fc = sensor_service.layers[0]
archive_service = gis.content.get(archive_id)
archive_fc = archive_service.layers[0]
```

```python
# Header to submit with request
h = {"X-API-Key": api_key}
# Make the request to the API
purple_response = requests.get(api_url + "?fields=" + fields, headers=h)
# JSON-ify the response
purple_sensors = purple_response.json()
```

```python
sensor_fc.delete_features(where="objectid > 0")
```

```python
sensors_update = []
record_index = 0
for index, sensor in enumerate(purple_sensors["data"]):
    sensors_update.append({"geometry": {
        'y': sensor[3],
        'x': sensor[4]
        },"attributes":{
        "sensor_index": sensor[0],
        "name": string_handler(sensor[1]),      
        "location_type": sensor[2],
        "latitude": sensor[3],
        "longitude": sensor[4],
        "altitude": none_handler(sensor[5]),
        "last_seen": sensor[6],
        "channel_state": sensor[7],
        "channel_flags": sensor[8],
        "confidence": sensor[9],
        "pm1": none_handler(sensor[10]),
        "pm25": none_handler(sensor[11]),
        "pm25_10min": none_handler(sensor[12]),
        "pm25_30min": none_handler(sensor[13]),
        "pm25_60min": none_handler(sensor[14]),
        "pm25_6h": none_handler(sensor[15]),
        "pm25_24h": none_handler(sensor[16]),
        "pm25_1w": none_handler(sensor[17]),
        "pm10": none_handler(sensor[18]),
        "count_03um": none_handler(sensor[19]),
        "count_05um": none_handler(sensor[20]),
        "count_1um": none_handler(sensor[21]),
        "count_25um": none_handler(sensor[22]),
        "count_5um": none_handler(sensor[23]),
        "count_10um": none_handler(sensor[24]),
        "humidity": none_handler(sensor[25]),
        "temperature": none_handler(sensor[26]),
        "pressure": none_handler(sensor[27]),
        "voc": none_handler(sensor[28]),
        "ozone": none_handler(sensor[29]),
        "analog_input": sensor[30]
    }})
    if(record_index <= 500):
        record_index = record_index + 1
    else:
        sensor_fc.edit_features(adds= sensors_update)
        archive_fc.edit_features(adds = sensors_update)
        record_index = 0
        sensors_update = []
    #print(index)
```
